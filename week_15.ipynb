#### `ì‹¤ìŠµ1`
readline íŒŒì¼ ì½ëŠ”ë°, 
while ì‚¬ìš©í•´ì„œ í•œì¤„ì”© ì½ê³  strip()í•´ì„œ ë¦¬ìŠ¤íŠ¸ë¡œ ë°˜í™˜í•˜ëŠ” í•¨ìˆ˜ ë§Œë“¤ê¸°

```python
def make_readlines(file):
    sent = []
    with open(file, 'r') as f:
        while content := f.readline().strip():
            sent.append(content)
            
    return sent

make_readlines('text.txt')
```
