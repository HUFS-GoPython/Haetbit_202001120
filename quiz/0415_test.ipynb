#### 1. X = [3, 1, -2, 6, 7]인 리스트에 대해 prefix sum S는 다음과 같이 정의된다.
- S[i] = X[0] + X[1] + ... + X[i]
- 즉, S = [3, 4, 2, 8, 15]가 된다.
이중 for loop를 사용하여 prefixSum 함수를 구현하고 함수의 수행시간을 구하시오.

[힌트]
실행시간을 측정하기 위해서는 아래와 같은 함수를 사용한다.
```import time
time.process.time() # 이 함수는 현재 clock 시간을 알려준다.
```
시간측정예시)
함수 f()의 시간을 측정하고 싶을 때,
```import time
before = time.process.time() # 현재 시간을 얻는다.
f() # 함수f 호출
after = time.process.time() # 현재 시간을 얻는다.
print(after - before) # 함수 호출 전과 후의 시간 차이
```

````python
import time

def prefix_sum(X):
    S = []
    for i in range(len(X)):
        num = 0
        for j in range(i + 1):
            num += X[j]
        S.append(num)
    return S


X = [3, 1, -2, 6, 7]

before = time.process_time()
prefix_sum(X)
after = time.process_time()

print(after - before)
````

#### 2. 다음은 n이 각각 1, 2, 3, 4, 5, 6일 때의 삼각수를 그림으로 나타낸 것이다.



양의 정수가 세 삼각수의 합으로 나타낼 수 있는지를 구하는 프로그램을 작성하시오.
- 입력: n은 1이상 1000이하의 양의 정수
- 출력: n을 세 수의 합으로 나타낼 수 있으면 이 세 삼각수를 오름차순으로 출력, 세 수의 합으로 나타낼 수 없으면 -1을 출력. (단, 해가 여러 개 있을 경우 세 수의 최솟값이 가장 작은 것출력)

[입출력 예시]

입력1) 10   출력1) 1 3 6
입력2) 6    출력2) -1

(힌트) n번째 삼각수 = n * (n + 1) / 2

````python
n = int(input())

# 44번째 삼각수: 990, 45번째 삼각수: 1035
triangle_num = [
    num * (num + 1) // 2
    for num in range(1, 45)
]

answer = False
for a in triangle_num:
    for b in triangle_num:
        for c in triangle_num:
            if (a + b + c) == n:
                nums = [a, b, c]
                nums.sort()

                answer = True
                break

        if answer:
            break

    if answer:
        break

if answer:
    for num in nums:
        print(num, end=" ")
else:
    print(-1)
````

#### 4. 소인수분해 (영어: prime factorization, integer factorization)는 1보다 큰 자연수를 소인수(소수인 인수)들만의 곱으로 나타내는 것 또는 합성수를 소수의 곱으로 나타내는 방법을 말한다.

예: 15 = 1 * 3 * 5 (소인수 = [1, 3, 5])
8 = 2 * 2 * 2 (소인수 = [1, 2])

무작위 수 n이 주어졌을 때 n의 소인수를 출력하는 함수를 만드시오.

단, 위의 예에서 8과 같은 경우는 2가 한 번만 출력되어야 함.
- n의 무작위 생성을 위해 다음의 코드를 참고하세요.
```
import random
random.randint(1, 300)
```

````python
import random
n = random.randint(1, 300)

def factorization(n):
    prime_list = []

    d = 2
    while d <= n:
        if n % d == 0:
            if not d in prime_list:
                prime_list.append(d)
            n = n / d
        else:
            d = d + 1

    return prime_list

print(f'{n}의 소인수는 {factorization(n)}입니다.')
````

#### 5. 통계학에서 N개의 수를 대표하는 기본 통계값에는 다음과 같은 것들이 있다. 단, N은 홀수라고 가정하자.
1. 산술평균: N개의 수들의 합을 N으로 나눈 값
2. 중앙값: N개의 수들의 합을 N으로 나눈 값
3. 최빈값: N개의 수들 중 가장 많이 나타나는 값
4. 범위: N개의 수들 중 최댓값과 최솟값의 차이
N개의 수가 주어졌을 때, 네 가지 기본 통계값을 구하는 프로그램을 작성하시오.

입력: 첫째 줄에 수의 개수 N(1이상 500,000이하)이 주어진다. 단, N은 홀수이다. 그 다음 N개의 줄에는 정수들이 주어진다. 입력되는 정수의 절댓값은 4,000을 넘지 않는다.

출력: 

첫째 줄에는 산술평균을 출력한다. 소수점 이하 첫째 자리에서 반올림한 값을 출력한다.

둘째 줄에는 중앙값을 출력한다.

셋째 줄에는 최빈값을 출력한다. 여러 개 있을 때에는 최빈값 중 가장 작은 값을 출력한다.

넷째 줄에는 범위를 출력한다. (힌트: 범위 = 최댓값 - 최솟값)

[입출력 예시]

______입력)   ____________  출력)
``` 
    5                2
    1                2
    3                1
    8                10
    -2
    2
```

````python
def nums_mean(lst):
    sum = 0
    for n in lst:
        sum += n

    return sum / len(lst)


def nums_mid(lst):
    lst.sort()

    return lst[len(lst)//2]


def nums_mode(lst):
    nums_set = set(lst)
    max = 0
    mode_num = 0
    for num in nums_set:
        if lst.count(num) > max:
            max = lst.count(num)
            mode_num = num

    return mode_num


def nums_range(lst):
    max_num = max(lst)
    min_num = min(lst)

    return max_num - min_num


n = int(input())
nums = [
    int(input())
    for _ in range(n)
]

print()
print(f'산술평균: {round(nums_mean(nums))}')
print(f'중앙값: {nums_mid(nums)}')
print(f'최빈값: {nums_mode(nums)}')
print(f'범위: {nums_range(nums)}')
````
